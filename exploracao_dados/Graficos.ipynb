```python
import matplotlib.pyplot as plt

ps = []
rs = []

ps_ob = []
rs_ob = []

ps_op = []
rs_op = []
for i in range(1,8):
    teste = pd.read_csv(f'output/desc_cliquesmax_{i}.txt')
    ps.append(teste['PRECISION'][1])
    rs.append(teste['RECALL'][1])
    ps_ob.append(teste['PRECISION_OB'][1])
    rs_ob.append(teste['RECALL_OB'][1])
    ps_op.append(teste['PRECISION_OP'][1])
    rs_op.append(teste['RECALL_OP'][1])

    print(f"{i} - P: {teste['PRECISION'][1]}")
    print(f"R: {teste['RECALL'][1]}")

fig, ax = plt.subplots()
ax.set_xlim([0,1])
ax.set_ylim([0,1])
ax.scatter(rs, ps, c=['#E6B0AA','#D98880','#CD6155','#C0392B','#A93226','#7B241C','#641E16'])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Cliquesmax')
plt.savefig('output/graficos/cliquesmax.png')
plt.show()
fig, ax = plt.subplots()
ax.set_xlim([0,1])
ax.set_ylim([0,1])
ax.scatter(rs_ob, ps_ob, c=['#E6B0AA','#D98880','#CD6155','#C0392B','#A93226','#7B241C','#641E16'])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Cliquesmax Obrigatorias')
plt.savefig('output/graficos/cliquesmax_ob.png')
plt.show()
fig, ax = plt.subplots()
ax.set_xlim([0,1])
ax.set_ylim([0,1])
ax.scatter(rs_op, ps_op,  c=['#E6B0AA','#D98880','#CD6155','#C0392B','#A93226','#7B241C','#641E16'])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Cliquesmax Optativas')
plt.savefig('output/graficos/cliquesmax_opt.png')
plt.show()
```
