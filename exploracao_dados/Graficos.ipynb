```python

dados_formados_path = 'input/dados_formados.csv'
dados_formados = pd.read_csv(dados_formados_path, delimiter=';')
def turn_to_date(x : str):
    try:
        x = int(x.split('/')[2][:2])
    except:
        #print(x)
        x = 0
    return x
dados_formados = dados_formados[dados_formados['INGRESSO'].map(turn_to_date) > 15]
```

```python
def create_array_percurso(id_estudante : int) -> list:
    periodo_atual = False
    ano_atual = False
    linha_atual = []
    todas_linhas = []
    for row in dados_formados[dados_formados['ID_ANONIMO'] == id_estudante].sort_values(by=['ANO','PERIODO.1']).iterrows():
        linha = row[1]
        if periodo_atual != linha['PERIODO.1'] or ano_atual != linha['ANO']:
            periodo_atual = linha['PERIODO.1']
            ano_atual = linha['ANO']
            if len(linha_atual) > 0:
                todas_linhas.append(linha_atual)
                linha_atual = []
        linha_atual.append(linha)
        
    return todas_linhas    
```

```python

random_student = (dados_formados['ID_ANONIMO'].sample().iloc[0])

dados_estudante = create_array_percurso(random_student)
G = nx.DiGraph()

for i, line in enumerate(dados_estudante):
    for disc in line:
        G.add_node(f'{disc["CODIGO"]}/{disc["ANO"]}-{disc["PERIODO.1"]}', trilha=disciplinas[disc['CODIGO']].trilha if disc['CODIGO'] in disciplinas.keys() else 'undef')
        if i > 0:
            for disc_ant in dados_estudante[i-1]:
                G.add_edge(f'{disc_ant["CODIGO"]}/{disc_ant["ANO"]}-{disc_ant["PERIODO.1"]}',f'{disc["CODIGO"]}/{disc["ANO"]}-{disc["PERIODO.1"]}')

import igraph as ig
H = ig.Graph.from_networkx(G)
ig.plot(H, bbox=(0, 0, 600, 400),
        vertex_color=[cores[trilha] for trilha in H.vs['trilha']])
```
