```python
disciplinas = []
with open('input/dependencias.txt') as f:
    for line in f:
        curr = line.split(';')
        if 'P8' in curr:
            curr = curr[1].split('\n')[0]
            disciplinas.append(curr)
```

```python
disciplinas_obrigatorias = ['GE70D', 'EEC31', 'CSS30','EEX23']
disciplinas = np.array(disciplinas)
disciplinas = disciplinas[[disc not in disciplinas_obrigatorias for disc in disciplinas]]
```

```python
dados_formados_path = 'input/dados_formados.csv'
dados_formados = pd.read_csv(dados_formados_path, delimiter=';')

dados_formados = dados_formados[[int(x[3].split('/')[2]) >= 14 for x in dados_formados.values]]
#dados_formados = dados_formados[[float(str(x[7]).replace(',','.')) >= 6 for x in dados_formados.values]]
dados_formados = dados_formados[[x[4] != 'ES70N' or x[5] > 2017 for x in dados_formados.values]]
dados_formados = dados_formados[[x[4] != 'FI70D' or x[5] > 2017 for x in dados_formados.values]]
dados_formados = dados_formados[[x[4] != 'FI70A' or x[5] > 2017 for x in dados_formados.values]]
dados_formados = dados_formados[[x[4] != 'GE70F' or x[5] > 2017 for x in dados_formados.values]]
dados_formados = dados_formados[dados_formados.CODIGO.isin(disciplinas)]
```

```python
veses_cursada = dados_formados[dados_formados.CODIGO.isin(disciplinas)].CODIGO.value_counts()

codigos = veses_cursada.keys()[:]
valores = veses_cursada.values[:]

plt.bar(range(len(codigos)),valores)
#plt.title("Vezes cursadas por curso")
plt.ylabel("Vezes cursadas")
plt.xlabel("Disciplina")
#plt.xticks(range(len(codigos)),codigos)

plt.savefig('output/analise_exp/vezescursada_formadosoptativas.png')
plt.show()
print(veses_cursada.head().to_string())

total_alunos = dados_formados.ID_ANONIMO.nunique()
n_disciplinas_cursadas = np.sum(veses_cursada, axis=0)
carga_p_aluno = np.sum(dados_formados.CH, axis=0)/total_alunos

print('\nTotal de alunos: ' + str(total_alunos))
print('\nTotal de disciplinas cursadas: ' + str(n_disciplinas_cursadas))

print('\nDisciplinas por aluno: ' + str(n_disciplinas_cursadas/total_alunos))
print('\nCarga horaria por aluno: ' + str(carga_p_aluno))
```

```python
from collections import defaultdict

trilhas_alunos = []
for student in dados_formados['ID_ANONIMO'].unique():
    trilhas = defaultdict(lambda: 0)
    for _, disc in dados_formados[dados_formados['ID_ANONIMO'] == student].iterrows():
        if float(disc['NOTA'].replace(',','.')) >= 6:
            trilhas[disciplinas[disc['CODIGO']].trilha] += disciplinas[disc['CODIGO']].ch
    trilhas = {k: v for k, v in sorted(trilhas.items(), key = lambda item: -item[1])}
    trilhas_alunos.append(trilhas)
```

```python
histograma_trilhas = defaultdict(lambda: 0)
for item in trilhas_alunos:
    for key in item:
        if item[key] > 60:
            histograma_trilhas[key] += 1
```

```python
histograma_trilhas = {k: v for k, v in sorted(histograma_trilhas.items(), key = lambda item: -item[1])}
histograma_trilhas
```

```python
ch_trilhas = defaultdict(lambda: 0)
for item in trilhas_alunos:
    for key in item:
        if item[key] > 60:
            ch_trilhas[key] += item[key]
```

```python
ch_trilhas = {k: v for k, v in sorted(ch_trilhas.items(), key = lambda item: -item[1])}
ch_trilhas
```

```python
ch_todas_trilhas = {key : 0 for key in trilhas}
for item in trilhas_alunos:
    for key in item:
        ch_todas_trilhas[key] += item[key]
ch_todas_trilhas = {k: v for k, v in sorted(ch_todas_trilhas.items(), key=lambda item: -item[1])}
```
