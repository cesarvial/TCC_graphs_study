Esta rede representa as interações dos personagens do romance de Victor Hugo.\
Compare alguns valores fornecidos pelo NetworkX com os disponíveis no Gephi

```python
import matplotlib.pyplot as plt
import networkx as nx
import csv

student = []
with open('dados_historicos_alunos_de_EngComputacaoCuritiba formados.csv', 'r') as csvfile:   
    reader = csv.DictReader(csvfile, delimiter=';')
    for row in reader:
        if (row["ID_ANONIMO"] == "13236"):
            student.append(row)
print(student)
```

```python
G = nx.Graph()

# disciplinas 
for line in student:
    G.add_node(line["CODIGO"] + "/" + line["ANO"] + "/" + line["PERIODO"])
G.nodes()
```

```python
print(G.number_of_nodes())
```

```python
# todas as dependencias entre disciplinas
with open("dependencias.txt") as file:
    for line in file:
        G.add_edge(line.rstrip().split(";")[0], line.rstrip().split(";")[1])
G.edges()
```

```python
print(G.number_of_edges())
```

```python
nx.draw(G, with_labels=True, arrows=True, node_size=50, font_size=5)
```

```python
G.degree()
```

```python
dicDegree = dict(G.degree() )
valores = sorted(set(dicDegree.values()))
valores

hist = [list(dicDegree.values()).count(x) for x in valores]

plt.plot(valores, hist, 'ro-') 

plt.xlabel('Grau')
plt.ylabel('Numero de nós')
```

```python
deg=dict(G.degree())
soma = 0
for i in deg.values():
    soma += i
aver = soma/len(deg)
print(aver)
print(nx.average_clustering(G))
print(nx.average_clustering(G, count_zeros=False))
print(nx.diameter(G))
#print(nx.clustering(G))
clus=dict(nx.clustering(G))

# Gephi: average clustering coefficient for nodes with degree > 1 ()
soma = 0
c = 0
for i in range(len(deg)):
    if list(deg.values())[i] > 1:
     soma += list(clus.values())[i]
     c +=1
clus_aver = soma/c;
print(clus_aver)
```
