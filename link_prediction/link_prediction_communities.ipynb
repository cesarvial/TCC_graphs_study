```python
# as comunidades são listas de ids de alunos, logo é fácil buscar os alunos de cada comunidade
print(comunidades)
# busca o estudante pelo id
def get_student_by_id(id):
    for e in estudantes:
        if (e.id_anonimo == id):
            return e.id_anonimo, e.trilhas, e.disciplinas
```

```python
# Define uma estrutura de comunidades que possui uma lista de ids (os alunos) e uma lista de trilhas (ordenada e com o número de estudantes)
comunidades_trilhas = []
for c in comunidades:
    comm = {"estudantes":[],
            "trilhas":[]}
    comm["estudantes"] = c
    em_trilha = []
    for trilha in trilhas:
        est_in_trilha = 0
        for e_id in range(len(c)-1):
            e_trilhas = get_student_by_id(c[e_id])[1]
            if (trilha[0] in e_trilhas):
                est_in_trilha += 1
        if (est_in_trilha > 0):
            em_trilha.append([trilha[0], est_in_trilha])
    comm["trilhas"] = sorted(em_trilha, key=lambda t: t[1], reverse=True)
    comunidades_trilhas.append(comm)
comunidades_trilhas
```

# Implementação da predição de links com dados de comunidade

```python
# Funções para dado um estudante e uma disciplina, 
# verificar qual o peso extra que deve ser adicionado na predição de links para tal
'''
    Primeiramente será verificado quais as disicplinas optativas que o aluno já fez.
    Tendo isso, iremos procurar qual a comunidade (ou clique) que possui as trilhas dessas disciplinas como majoritárias.
    Com a comunidade, iremos atribuir um peso para todas as disciplinas das trilhas de tal comunidade/clique.
    Cálculo do peso:
        Utilizando a porcentagem de alunos que finalizaram a trilha de cada disciplina. Inicialmente, todas as disciplinas de uma trilha terão o mesmo peso
        Depois, podemos verificar dentro de cada trilha, quais foram as disciplinas mais realizadas, diminuindo o peso de uma disciplina que aparece pouco
                                                                                                     ou até zerando de uma que não aparece.
        Caso a disciplina seja obrigatória, ele deve ser adicionada de qualquer maneira.
    No final, teremos uma estrutura que relaciona o estudante com o peso "extra" que foi calculado para cada disciplina optativa
'''

'''
    Com a estrutura do aluno finalizada, podemos apenas buscar o peso extra de cada disciplina.
    A função depende dos vértices u, v e do aluno, para verificar se ele já fez alguma das duas disciplinas 
    E se a outra deve ser adicionada
'''
```

```python
# Funções de predição de link re-implementadas utilizando 
from math import log
import networkx as nx
from networkx.utils import not_implemented_for

# função para link prediction pego do networkx
'''
    Acredito que o peso extra possa ser colocado aqui, e não precise ser em cada alg em específico
    A função simplesmente será chamada junto da func(u, v), e seus resultados serão somados
'''
def apply_prediction(G, func, ebunch=None):
    if ebunch is None:
        ebunch = nx.non_edges(G)
    return ((u, v, func(u, v)) for u, v in ebunch)

# ---------------------------------------------------------------------------------------------------------------------
# Jaccard
@not_implemented_for("directed")
@not_implemented_for("multigraph")
def jaccard_coefficient(G, ebunch=None):
    def predict(u, v):
        union_size = len(set(G[u]) | set(G[v]))
        if union_size == 0:
            return 0
        return len(list(nx.common_neighbors(G, u, v))) / union_size
    return apply_prediction(G, predict, ebunch)

# ---------------------------------------------------------------------------------------------------------------------
# Preferential Attachment
@not_implemented_for("directed")
@not_implemented_for("multigraph")
def preferential_attachment(G, ebunch=None):
    def predict(u, v):
        return G.degree(u) * G.degree(v)
    return apply_prediction(G, predict, ebunch)

# ---------------------------------------------------------------------------------------------------------------------
# Resource Allocation
@not_implemented_for("directed")
@not_implemented_for("multigraph")
def resource_allocation_index(G, ebunch=None):
    def predict(u, v):
        return sum(1 / G.degree(w) for w in nx.common_neighbors(G, u, v))
    return apply_prediction(G, predict, ebunch)

# ---------------------------------------------------------------------------------------------------------------------
# Common Neighbor Centrality
@not_implemented_for("directed")
@not_implemented_for("multigraph")
def common_neighbor_centrality(G, ebunch=None, alpha=0.8):
    # When alpha == 1, the CCPA score simplifies to the number of common neighbors.
    if alpha == 1:

        def predict(u, v):
            if u == v:
                raise nx.NetworkXAlgorithmError("Self links are not supported")

            return sum(1 for _ in nx.common_neighbors(G, u, v))
    else:
        spl = dict(nx.shortest_path_length(G))
        inf = float("inf")
        def predict(u, v):
            if u == v:
                raise nx.NetworkXAlgorithmError("Self links are not supported")
            path_len = spl[u].get(v, inf)

            return alpha * sum(1 for _ in nx.common_neighbors(G, u, v)) + (
                1 - alpha
            ) * (G.number_of_nodes() / path_len)
    return apply_prediction(G, predict, ebunch)

# ---------------------------------------------------------------------------------------------------------------------
# Adamic-Adar
@not_implemented_for("directed")
@not_implemented_for("multigraph")
def adamic_adar_index(G, ebunch=None):
    def predict(u, v):
        return sum(1 / log(G.degree(w)) for w in nx.common_neighbors(G, u, v))
    return apply_prediction(G, predict, ebunch)
```
