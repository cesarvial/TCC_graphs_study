```python
# as comunidades so listas de ids de alunos, logo  fcil buscar os alunos de cada comunidade
print(comunidades)
# busca o estudante pelo id
def get_student_by_id(id):
    for e in estudantes:
        if (e.id_anonimo == id):
            return e.id_anonimo, e.trilhas, e.disciplinas
```

```python
# Define uma estrutura de comunidades que possui uma lista de ids (os alunos) e uma lista de trilhas (ordenada e com o nmero de estudantes)
comunidades_trilhas = []
for c in comunidades:
    comm = {"estudantes":[],
            "trilhas":[]}
    comm["estudantes"] = c
    em_trilha = []
    for trilha in trilhas:
        est_in_trilha = 0
        for e_id in range(len(c)-1):
            e_trilhas = get_student_by_id(c[e_id])[1]
            if (trilha[0] in e_trilhas):
                est_in_trilha += 1
        if (est_in_trilha > 0):
            em_trilha.append([trilha[0], est_in_trilha])
    comm["trilhas"] = sorted(em_trilha, key=lambda t: t[1], reverse=True)
    comunidades_trilhas.append(comm)
comunidades_trilhas
```

```python
# os cliques so um pouco mais chatos de buscar, j que eles so uma lista de posio do vrtice
print(cliques)
# passa como parmetro a posio do vrtice no grafo
def get_student_by_clique_pos(pos):
    return get_student_by_id(H.vs[pos]['_nx_name'])
```

```python
# Lista de trilhas com suas disciplinas
trilhas_map = []
with open("../texts/trilhas.txt") as f:
    for line in f.readlines():
        trilhas_map.append(line.split("\n")[0].split(";"))

# Define uma estrutura de cliques que possui uma lista de ids (os alunos) e uma lista de trilhas (ordenada e com nmero de alunos)
cliques_trilhas = []
for clique in cliques:
    cliq = {"estudantes":[],
            "trilhas":[],
            "disciplinas":[]}
    em_trilha = []  
    all_dis = [] 
    n_estudantes = 0    
    for c in clique:
        e_id = get_student_by_clique_pos(c)[0]
        n_estudantes += 1
        cliq["estudantes"].append(e_id)
        e = [e for e in estudantes if e.id_anonimo == e_id][0]
        # Adiciona todas as disciplinas que fazem parte das trilhas do estudante dentro da lista de discplinas do clique
        for t in e.trilhas:
            trilha = [tri for tri in trilhas_map if tri[0] == t][0]
            for d in e.disciplinas:
                if (d in trilha):
                    all_dis.append(d)
    for trilha in trilhas_map:
        est_in_trilha = 0
        for c in clique:
            e_trilhas = get_student_by_clique_pos(c)[1]
            if (trilha[0] in e_trilhas):
                est_in_trilha += 1
        if (est_in_trilha > 0):
            em_trilha.append([trilha[0], est_in_trilha])
    cliq["trilhas"] = sorted(em_trilha, key=lambda t: t[1], reverse=True)
    n_apperances = len(all_dis)
    # a 1a posio da disciplina mostra o nmero total de optativas realizadas
    cliq["disciplinas"].append(n_apperances)
    for d in all_dis:
        if d != "-":
            n_times = all_dis.count(d)
            # disciplina seguido pelo n de alunos que a realizou no clique
            cliq["disciplinas"].append([d, n_times])
            # evita inseres repetidas no futuro
            for idx, item in enumerate(all_dis):
                if item == d:
                    all_dis[idx] = '-'
        
    #print(cliq)
    cliques_trilhas.append(cliq)
cliques_trilhas
    
```

# Implementao da predio de links com dados de comunidade

```python
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib import colormaps
import sys
from itertools import combinations
from datetime import datetime
import pandas as pd
import numpy as np
sys.path.append('../utils')
from mapa_estudantes import get_historico_completo, gerar_estudantes, get_disciplinas_matriz_844
import igraph as ig
import random

# Funes para dado um estudante e uma disciplina, 
# verificar qual o peso extra que deve ser adicionado na predio de links para tal
'''
    Primeiramente ser verificado quais as disicplinas optativas que o aluno j fez.
    Tendo isso, iremos procurar qual a comunidade (ou clique) que possui as trilhas dessas disciplinas como majoritrias.
    Com a comunidade, iremos atribuir um peso para todas as disciplinas das trilhas de tal comunidade/clique.
    Clculo do peso:
        Utilizando a porcentagem de alunos que finalizaram a trilha de cada disciplina. Inicialmente, todas as disciplinas de uma trilha tero o mesmo peso
        Depois, podemos verificar dentro de cada trilha, quais foram as disciplinas mais realizadas, diminuindo o peso de uma disciplina que aparece pouco
                                                                                                     ou at zerando de uma que no aparece.
        Caso a disciplina seja obrigatria, ele deve ser adicionada de qualquer maneira.
    No final, teremos uma estrutura que relaciona o estudante com o peso "extra" que foi calculado para cada disciplina optativa
'''

# Dado um Ra, cria o grafo do caminho completo do estuante
def create_full_student_graph(ra):
    # Cria um grafo do percurso completo do aluno
    # Aluno com todas as disciplinas
    G_aluno_curso = nx.Graph()
    disciplinas = df_844_curso["CODIGO"].unique()
    G_aluno_curso.add_nodes_from(disciplinas)

    # Exibir o nmero de disciplinas
    df_844_curso_aluno = df_844_curso[df_844_curso["ID_ANONIMO"] == ra]

    for data in df_844_curso_aluno.groupby(["ID_ANONIMO"]):
        dados_estudante = data[1]
        disciplinas = (
            dados_estudante.sort_values(by=["SEMESTRE"])
            .groupby(["CODIGO"])
            .agg({"SEMESTRE": "last"})
            .sort_values(by=["SEMESTRE"])
            .itertuples()
        )
        for disc1, disc2 in combinations(disciplinas, 2):
            if semestres_consecutivos(disc1.SEMESTRE, disc2.SEMESTRE):
                if G_aluno_curso.has_edge(disc1.Index, disc2.Index):
                    G_aluno_curso[disc1.Index][disc2.Index]["weight"] += 1
                else:
                    G_aluno_curso.add_edge(disc1.Index, disc2.Index, weight=1)

    return G_aluno_curso

# Dado um grafo e o nome de um vrtice, remove todas as arestas desse vrtice
# Usado para remover disciplinas especficas (como optativas para teste)
def remove_all_edges_from_node(g, node_name):
    edge_subset = [e for e in g.edges() if (e[0] == node_name or e[1] == node_name)]
    new_g = g.copy()
    new_g.remove_edges_from(edge_subset)
    return new_g

# Dado um grafo, remove todas as arestas de 20% dos ns de um grafo
# No parece ser muito til, com essa funo os grafos perdem um pouco o sentido
def remove_random_nodes(g):
    # Remove as arestas de 20% dos vrtices
    #  interessante pegar apenas vrtices que tem pelo menos alguma aresta
    nodes_with_edges = [node for node in g.nodes() if g.degree[node] > 0]
    proportion_edges = 0.2
    node_subset = list(random.sample(
        list(nodes_with_edges), int(proportion_edges * len(nodes_with_edges))
    ))
    print("Nodes removed: ", node_subset)
    # Cria uma cpia do grafo e remove as arestas
    new_g = g.copy()
    for node in node_subset:
        new_g = remove_all_edges_from_node(new_g, node)
    
    return new_g

# Dado um grafo de estudante, remove todas as arestas de todos menos 1 n de optativa
def maintain_one_opt_node(g, n_maintained=1):
    # Busca quais so os vrtices de disciplinas optativas
    disciplinas = df_844_optativas["CODIGO"].unique()
    nodes_with_edges = [node for node in g.nodes() if g.degree[node] > 0]
    opt_nodes_with_edges = [node for node in disciplinas
                            if node in nodes_with_edges]
    random.shuffle(opt_nodes_with_edges)
    print("Opts:", opt_nodes_with_edges)
    # mantm n_maintained disicplinas, logo retira n_maintained da lista de ns que sero removidos
    for i in range(n_maintained):
        opt_nodes_with_edges.pop()
    print("Removed Opts:", opt_nodes_with_edges)
    new_g = g.copy()
    for node in opt_nodes_with_edges:
        new_g = remove_all_edges_from_node(new_g, node)
    return new_g

# Dado uma disciplina, retorna uma lista de cliques nos quais ela pertence
def get_course_cliques(codigo):
    cs = []
    for clique in cliques_trilhas:
        for d in clique['disciplinas'][1:]:
            if codigo == d[0]:
                cs.append(clique)
                break
    print("CLIQUES:", cs)
    
    return cs

# Dado um grafo, retorna uma lista de disciplinas recomendadas com a sua probabilidade de ser realizada
def get_recommended_courses(g):
    # pega as disciplinas optativas existentes
    disciplinas = df_844_optativas["CODIGO"].unique()
    nodes_with_edges = [node for node in g.nodes() if g.degree[node] > 0]
    opt_nodes_with_edges = [node for node in disciplinas
                            if node in nodes_with_edges]
    print(opt_nodes_with_edges)
    # caso no tenham disciplinas optativas existentes, roda com base no interesse
    # TODO: no momento iremos apenas ignorar esse caso
    if (len(opt_nodes_with_edges) == 0):
        print("Sem disciplinas optativas j realizadas. Sem recomendaes.")
        return []
    # Caso tenham disciplinas, busca os cliques de cada uma delas
    e_cliques = []
    for node in opt_nodes_with_edges:
        e_cliques.append(get_course_cliques(node))
    all_rec_dis = []
    n_dis = 0
    # para buscar a recomendao divide a quantidade de vezes que uma disciplina apareceu em todos os cliques pela quantidade total de disciplinas
    # Conta o nmero total de disciplinas
    for clique in e_cliques:
        print(clique)
        n_dis += clique[0]['disciplinas'][0]
    # Adiciona todas as disicplinas em um vetor
    for clique in e_cliques:
        for d in clique[0]['disciplinas'][1:]:
            if d[0] not in opt_nodes_with_edges:
                for i in range(d[1]):
                    all_rec_dis.append(d[0])
    # Pega o valor das disciplinas
    rec_dis = []
    for d in all_rec_dis:
        if d != '-':
            nd = all_rec_dis.count(d)
            rec_dis.append([d, 10*(nd/n_dis)])
            all_rec_dis[:] = [x if x != d else '-' for x in all_rec_dis]
    
    return rec_dis


        
```
